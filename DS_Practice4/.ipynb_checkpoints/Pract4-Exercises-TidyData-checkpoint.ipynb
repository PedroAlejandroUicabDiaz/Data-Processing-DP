### Parte One - Key Concepts

***1. What is data tidying?***

The tidy data standard has been designed to facilitate initial exploration and analysis of the data, and to simplify the develoment of data analysis tools tha works well together. In few words  the "data tidying" is the process of structuring our data in a way that makes it easy to analyze and use.

***2. What is data structure?***

Data structure is a process through which data is stored and arranged in the disk of our computer or also memory storage, in away that the data can be easily used and manipulated in the future. Furthermore, it is am effective way of performing various operations ralted to data management. In addition, we can use differente data structures to solve certain problems.

***3. What is data semantic?***

It is like the way to interpret the real-world in variables, as we know a dataset is a collection of values(quantitative and qualitative data) and we can organised in two ways: variables and observation.

Variables contains all values that measure the same underlying attribute(like height, temperature, duration) and an observation contains all values measured on the same unit (like person, day or race) across attributes.

***4. What is Tidy Data? What are the three main characteristics of Tidy Data?***

Tidy data is a standard of mapping the meaning of a dataset to its structure.

**(1)** - Each variable forms a column

**(2)** - Each observation forms a row

**(3)** - Each type of observational unit form a table

***5. How can identified if data is messy or tidy?***

It depends on how rows, columns and tables are mached up with observations , variables and types. Also a messy data is any another arrangement of 
the data. 

***6. What are the five most common problems in messy datasets?***

**(1)** - Column headers are values, not variable names

**(2)** - Multiple varibales are stored in one column 

**(3)** - Variables are stored in both rows and columns

**(4)** - Multiple types of the observational units are stored in the same table.

**(5)** - A single observational units is stored in multiple tables

***7. Choose two datasets from the Tidy Data Examples. Explain the identified error and how it can be cleaned in each case.***

```python
# Import the packes
import pandas as pd

# Load the csv file
religions_df = pd.read_csv("Religion/pew.csv")

# The first five rows to have an idea about the data frame
religions_df.head(5)
```

```python
# The last five rows
religions_df.tail(5)
```

```python
# Description about data frame
"""
We can see that our data frame has 18 rows and 11 columns, where our 
header has values too and not variables/names.

To sort the data frame we need to become the frecuency in rows instead of columns 
as the original version.

To be able to perform this operation we need the MELT function from pandas; Melth function is useful to massage a 
DataFrame into a format where one or more columns are identifier variables  
"""
religions_df.info()
```

```python
"""
After the use melt function from pandas we will able to re-organize our 
data frame so, we will have a tidy dataframe.
"""
# Using melt Function
religions_df = religions_df.melt(id_vars=["religion"],var_name=["income"],value_name="frequency")

# See the changes
religions_df.head()
```

### (2) - Multiple varibales are stored in one column

**Dataset: c_tb.csv**

```python
# Load the csv file
tuber_df = pd.read_csv("Tuberculosis/c_tb.csv")

# The first five rows to have an idea about the data frame
tuber_df.head(5)
```

```python
# The last five rows
tuber_df.tail(5)
```

```python
# Description about data frame
"""
We can see that our data frame has 5768 rows and 23 columns, where each columns contains two different variables, the 
sex and range age.

To sort the data frame we need to separate both variables into differents columns.

To be able to perform this operation we need the MELT function from pandas; Melth function is useful to massage a 
DataFrame into a format where one or more columns are identifier variables  
"""
tuber_df.info()
```

```python
# Columns
tuber_df.columns
```

```python
"""
Now that we could see the all comlumns the first step will be moving 
the data frame by NaN values, then that we have the variable 
demografic (sex/rangeage) into one column we will divide it in different
columns [age][sex].
"""

# For better data management we will change the names a bit, see the change

tuber_df.columns = ['iso2', 'year', 'total', 'm04', 'm514', 'm014', 'm1524', 'm2534', 'm3544', 'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534', 'f3544', 'f4554', 'f5564', 'f65', 'fu']

tuber_df.columns
```

```python
# Now we will remove the columns that we won't work

tuber_df=tuber_df.drop(['total','m04','m514','f04','f514'], axis=1)
```

```python
# Verify it 

tuber_df.columns
```

```python
# Now, we are going to use Melt function from pandas
tuber_df = tuber_df.melt(id_vars=["iso2","year"],var_name=["demographic"],value_name="cases")
```

```python
tuber_df.head()
```

```python
# So far, we are ready to divide demografic columns into different columns

tuber_df = (tuber_df.assign(sex = lambda x: x.demographic.str[0].astype(str), age = lambda x: x.demographic.str[1:].astype(str)).drop("demographic",axis=1))
```

```python
# Verify the change
tuber_df.head()
```

```python
# As you can see, we have various NaN values so, we need to clean it
tuber_df.update(pd.DataFrame({'age':[age[:2]+'-'+age[2:] 
                            if len(age) == 4 else (age) 
                            for age in tuber_df["age"]]}))
tuber_df = (tuber_df.replace(to_replace =["m","f","014","65","u"],value =["Male","Female","0-14","65+","unknown"]).dropna())
```

```python
# Verify it 
tuber_df.sample(10)
```

```python
tuber_df.head()
```

```python
# Now we have our data frame ready, so the last step is save it
#tuber_df.to_csv('Tuberculosis/c_tb_modified.csv')
```

### (3) - Variables are stored in both rows and columns

**Dataset: d_weather.csv**

```python
# Load the csv file
weather_df = pd.read_csv("Weather/d_weather.csv")

# The first five rows to have an idea about the data frame
weather_df.sample(5)
```

```python
# The last five rows
weather_df.tail(5)
```

```python
# Description about data frame
"""
We can see that our data frame has 9 rows and 12 columns, where our data frame has varaibles in columns and rows.
See that [element] belongs or refers to tempetures and [d1,d2,d3..] refers to days so, we must start sorting these columns.

To sort the data frame we need to use melt function to create a row for each record for the day variables.

To be able to perform this operation we need the MELT function from pandas; Melth function is useful to massage a 
DataFrame into a format where one or more columns are identifier variables  
"""
weather_df.info()
```

```python
# Melt function
weather_df=weather_df.melt(id_vars=["id","year","month","element"],var_name=["day"],value_name="temp")
weather_df.update(pd.DataFrame({"day":[day[1:] for day in weather_df["day"]]}))

# Verify the changes
weather_df.sample(5)
```

```python
"""
So far we did the first step, now we are going to divide the [element] column in two different columns, 
one for [tmac] and the another for [tmin]. Doing this we will be able to sort this data frame correctly.

To be able to perfom this operation we are going to use [ dataframe.pivot_table(parameters) ] method. That will help us to create 
a spreadsheet-style pivot table as a DataFrame.
"""
# Using pivot_table
weather_df = (weather_df.pivot_table(index = ["year","month","day","id"], columns = "element", values = "temp").reset_index().rename_axis(None, axis = 1))
```

```python
"""
As you can see, our table is already sorted, the day and tempeture variables now are right!
"""
# Verify it 
weather_df.head(5)
```

```python
# Now we have our data frame ready, so the last step is save it
#weather_df.to_csv('Weather/d_weather_modified.csv')
```

### (5) - A single observational units is stored in multiple tables

```python
# Load the csv files
Baby14_df = pd.read_csv("Babies_name/2014-baby-names-illinois.csv")
Baby15_df = pd.read_csv("Babies_name/2015-baby-names-illinois.csv")
```

```python
# The first five rows to have an idea about the data frame (2014)
Baby14_df.head(5)
```

```python
# The first five rows to have an idea about the data frame (2015)
Baby15_df.head(5)
```

```python
Baby14_df.info()
```

```python
Baby15_df.info()
```

```python
# Creating new column [year] for both datasets
Baby14_df["year"]="2014"
Baby15_df["year"]="2015"
```

```python
# Concatenating the datasets
Baby1415_df = pd.concat([Baby14_df, Baby15_df]).sort_values(by=['rank'])

# Set new index
(Baby1415_df.set_index('rank', inplace=True))
```

```python
# Verify it
Baby1415_df.head(5)
```

```python
# Re check the dataframe info
Baby1415_df.info()
```
